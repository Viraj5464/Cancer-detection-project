# Breast Cancer Prediction:

```python
import numpy as np
import pandas as pd
#importing libraries
```

```python
data = pd.read_csv("breast-cancer.csv")
#reading dataset
```

```python
data=data.iloc[: ,1:]
data
```

```python
data
```

```python
data[["radius_mean","texture_mean","perimeter_mean","area_mean"]] 
#These are the attributes should be converted into the range 0 to 1.
```

```python
from sklearn.preprocessing import MinMaxScaler,LabelEncoder
#importing minmax scaler and labelEncoder
```

```python
SS = MinMaxScaler()
LE = LabelEncoder()
```

```python
data[["radius_mean","texture_mean","perimeter_mean","area_mean"]] = SS.fit_transform(data[["radius_mean","texture_mean","perimeter_mean","area_mean"]])
#converting these attributes to the range of 0 and 1 by using minmaxscaler
```

```python
data
```

```python
data["diagnosis"].unique()
#m-malignant(harmful) b-benign(not harmful)
```

```python
data["diagnosis"] = LE.fit_transform(data["diagnosis"])
#converting dependent variable by using label encoder.
```

```python
data
```

```python
data.corr()
#displays the correlation values.
```

```python
from sklearn.model_selection import train_test_split
#model selection.
```

```python
X = data[["radius_mean","texture_mean","perimeter_mean","area_mean","smoothness_mean","compactness_mean","concavity_mean","concave points_mean","symmetry_mean","radius_se","perimeter_se","area_se","compactness_se","concavity_se","concave points_se","fractal_dimension_se","radius_worst","texture_worst","perimeter_worst","area_worst","smoothness_worst","compactness_worst","concavity_worst","concave points_worst","symmetry_worst","fractal_dimension_worst"]]
y = data.iloc[: , 0:1]
X
#assigning all the required independent variables to x.
#assigning dependent variable to y.
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=124)
#assigning values for training and testing.
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn import svm
#model building
```

```python
DTC = DecisionTreeClassifier()
LogReg = LogisticRegression()
NB = GaussianNB()
SVM = svm.SVC()
```

```python
DTC.fit(X_train ,  y_train)
LogReg.fit(X_train ,  y_train)
NB.fit(X_train ,  y_train)
SVM.fit(X_train ,  y_train)
```

```python
pred1 = DTC.predict(X_test)
pred2 = LogReg.predict(X_test)
pred3 = NB.predict(X_test)
pred4 = SVM.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
score1 = r2_score(pred1  , y_test)
```

```python
score2 = r2_score(pred2  , y_test)
```

```python
score3 = r2_score(pred3  , y_test)
```

```python
score4 = r2_score(pred4  , y_test)
```

```python
import matplotlib.pyplot as plt
 
data1 = {'DTC':score1, 'LOGR':score2, 'NB':score3,'SVM':score4}
courses = list(data1.keys())
values = list(data1.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(courses, values, color ='maroon',
        width = 0.4)
 
plt.xlabel("ALGORITHM")
plt.ylabel("ACCURACY SCORE")
plt.title("BREAST CANCER DETECTION - COMPARISION ")
plt.show()

print(" Decison Tree Algorithm Accuracy:",score1*100,"\n","Logistic Regression Algorithm Accuracy",score2*100,"\n","Naive Bayes Algorithm Accuracy",score3*100,"\n","SVM Algorithm Accuracy",score4*100)
#accuracy
```
